```python
from sympy import *
from sympy.abc import s
from sympy.physics import control
from sympy.physics.control.lti import TransferFunction

init_printing(use_unicode=True)
w0,k0,k1,k2,Q = symbols("w0 k0 k1 k2 Q", real=True, constant= True)
vi,vo,u0 = symbols("vi vo u0")
```

```python
u0 = vi*k0 - w0*vo/s + k1*s*vi 

vo = u0/s 

hexpr = vo/vi 

Hs = TransferFunction.from_rational_expression(hexpr,s)

display(Hs.expand())
```

```python




u1= -w0*vo + vi*k0/w0

u2 = -w0/Q*vo + w0*u1/s + k1*vi + k2*s*vi

vo = u2/s

Hs = vo/vi
```

```python
display(simplify(Hs))
```
